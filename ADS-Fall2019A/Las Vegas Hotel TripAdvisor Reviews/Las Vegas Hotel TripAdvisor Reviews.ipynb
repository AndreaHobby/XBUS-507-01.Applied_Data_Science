### Regression

```python
features = ['Period of stay','Traveler type','Pool','Gym','Tennis court','Spa','Casino',
            'Free internet','User continent','Member years','Review month']
getdummies = ['Period of stay','Traveler type','Pool','Gym','Tennis court','Spa','Casino',
      'Free internet','User continent','Review month']
```

```python
y = df['Score']
X = df[features]
```

```python
X = pd.concat([pd.get_dummies(df[getdummies]), df['Member years']], axis=1)
```

```python
pd.set_option('max_columns',500)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
coefficients = pd.concat([pd.DataFrame(X.columns),pd.DataFrame(np.transpose(lr.coef_))], axis = 1)
print(coefficients)
print(r2_score(y_test, y_pred))
```
